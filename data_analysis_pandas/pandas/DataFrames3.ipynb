# Filter with `query`

## Allows more Pythonic language inside query. Requires no spaces in column names

```python
bond = pd.read_csv("jamesbond.csv", index_col="Film")
bond = bond.sort_index()
bond.head(3)
```

```python
[column_name.replace(" ", "_") for column_name in bond.columns]
```

```python
bond.columns = [column_name.replace(" ", "_") for column_name in bond.columns]
```

```python
bond.head(3)
```

```python
bond.query("Actor == 'Sean Connery'")
```

```python
bond.query("Director == 'Terence Young'")
```

```python
bond.query("Actor != 'Roger Moore'")
```

```python
bond.query("Box_Office > 600")
```

```python
bond.query("Actor == 'Roger Moore' and Director == 'John Glen'")
```

```python
bond.query("Actor in ['Timothy Dalton', 'George Lazenby']")
```

```python
bond.query("Actor not in ['Sean Connery', 'Roger Moore']")
```

# `.apply()`

```python
bond = pd.read_csv("jamesbond.csv", index_col="Film")
bond = bond.sort_index()
bond.head(3)
```

```python
bond.head()
```

```python
def convert_to_string_and_add_million(number):
    return str(number) + " MILLION!"

bond["Box Office"] = bond["Box Office"].apply(convert_to_string_and_add_million)
```

```python
bond
```

```python
bond["Budget"] = bond["Budget"].apply(convert_to_string_and_add_million)
bond
```

```python
bond = pd.read_csv("jamesbond.csv", index_col="Film")
bond = bond.sort_index()
bond.head(3)
```

```python
columns = ["Box Office", "Budget", "Bond Actor Salary"]
for col in columns: 
    bond[col] = bond[col].apply(convert_to_string_and_add_million)
bond.head(3)
```

# `.apply()` with Rows

```python
bond = pd.read_csv("jamesbond.csv", index_col="Film")
bond = bond.sort_index()
bond.head(3)
```

```python
def good_movie(row): 
    
    actor = row[1]
    budget = row[4]
    
    if actor == "Pierce Brosnan":
        return "The Best"
    elif actor == "Roger Moore" and budget > 40:
        return "Enjoyable"
    else:
        return "Idk"

bond.apply(good_movie, axis="columns")
```
